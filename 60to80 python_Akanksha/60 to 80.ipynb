62. what % of all patients with an outcome of "Alive" died within 28 days. What was the mean platelet volume among this group?

```python
alive_patients= df_Hospitalization_Discharge[df_Hospitalization_Discharge['outcome_during_hospitalization'] == 'Alive']
alive_patients
```

```python
patients_died_within28days=alive_patients[alive_patients['death_within_28_days'] == '1']
patients_died_within28days
a = len(patients_died_within28days)/len(df_Hospitalization_Discharge) 
percentage_died_within_28_days =a * 100
print(f"Percentage of 'Alive' patients who died within 28 days: {percentage_died_within_28_days:.2f}%")
```

```python
mean_platelet_volume = alive_patients[patients_died_within28days]['mean_platelet_volume'].mean()
```
