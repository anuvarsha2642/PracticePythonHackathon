```python
import pandas as pd

# Sample DataFrame
data = {
    'Date': ['2024-01-01', '2024-01-01', '2024-01-01', '2024-01-01', '2024-02-01', '2024-02-01', '2024-02-01', '2024-02-01'],
    'Region': ['East', 'East', 'West', 'West', 'East', 'East', 'West', 'West'],
    'Product': ['Laptop', 'Smartphone', 'Laptop', 'Smartphone', 'Laptop', 'Smartphone', 'Laptop', 'Smartphone'],
    'Sales': [100, 200, 150, 180, 120, 220, 130, 190]
}

df = pd.DataFrame(data)

# Creating a pivot table
pivot_table = pd.pivot_table(df, values='Sales', index=['Date', 'Region'], columns='Product', aggfunc='sum')

print(pivot_table)
```
